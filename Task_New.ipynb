```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
a=[]  # Will store doctor names
h=[]  #  Will store clinic addresses

# Define a function to extract doctor names and clinic addresses from a webpage
def fun(webpage):
    soup=BeautifulSoup(webpage,'lxml')  # Create a BeautifulSoup object to parse the HTML

    b=soup.find_all('a',class_='doctor-card__name')  # Find all 'a' tags with the class 'doctor-card__name'

    # Iterate through the found 'a' tags and extract the text content (doctor names)
    for i in b:
      a.append(i.text.strip())  # Append the cleaned name to the 'a' list

    # Find all 'div' tags with the class 'doctor-card__clinic'
    k=soup.find_all('div',class_='doctor-card__clinic')
    
    # Iterate through the found 'div' tags and extract the text content (clinic addresses)
    for i in k:
      h.append(i.text.strip())   # Append the cleaned address to the 'h' list
        
    # Return both lists containing extracted data
    return a,h


# Loop through multiple pages of the website
for i in range(1,15):
    # Construct the URL for each page

    webpage=requests.get(f'https://us-uk.bookimed.com/doctors/country=germany/page={i}/',).text  # Fetch HTML content of the page
    y=fun(webpage)  # Call the 'fun' function to extract data from the page




    

    
    
```

```python
y=[]
p=[]
w=[]
# Define a function to extract doctor names and clinic addresses from clinicsoncall.com

def function(webpage):   # Create a BeautifulSoup object to parse the HTML
    soup=BeautifulSoup(webpage,'lxml')

    # Find all 'h2' tags with the class 'doctor-name'
    for i in soup.find_all('h2',class_='doctor-name'):
        y.append(i.text.strip())  # Append the cleaned doctor name to the 'y' list

    # Find all 'div' tags with the class 'doctor-info-text'

    for i in soup.find_all('div',class_='doctor-info-text'):
        p.append(i.text.strip())   # Append the cleaned text content to the 'p' 
    # Return both lists containing extracted data
    return y,p

# Set headers to mimic a regular browser request
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
# Loop through multiple pages of the website
for i in range(1,15):
    # Construct the URL for each page
    webpage=requests.get(f'https://clinicsoncall.com/en/doctors/country-germany/page/{i}/',headers=headers).text  # Fetch HTML content of the page
    x=function(webpage)   # Call the 'function' function to extract data from the page

#print(x[0])
# Extract the first line from each 'div' tag in the 'p' list
for i in x[1]:
    w.append(i.split('\n')[0])  # Append the first line (presumably the address) to the 'w' list


#w
```
